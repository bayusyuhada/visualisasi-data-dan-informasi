```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari CSV
data = pd.read_csv('/content/Jumlah Penduduk Menurut Kecamatan dan Jenis Kelamin di Kabupaten Lampung Selatan, 2020-2021.csv', skiprows=3)

# Tampilkan data asli
print(data.head())

# Mengatur nama kolom untuk kemudahan akses
data.columns = ['Kecamatan', '2020_Laki-Laki', '2021_Laki-Laki', '2020_Perempuan', '2021_Perempuan', '2020_Jumlah', '2021_Jumlah']

# Membuat visualisasi jumlah penduduk laki-laki dan perempuan di setiap kecamatan pada tahun 2020
plt.figure(figsize=(12, 6))
bar_width = 0.35
x = range(len(data['Kecamatan']))

plt.bar(x, data['2020_Laki-Laki'], width=bar_width, label='Laki-Laki 2020')
plt.bar(x, data['2020_Perempuan'], width=bar_width, bottom=data['2020_Laki-Laki'], label='Perempuan 2020')

plt.xlabel('Kecamatan')
plt.ylabel('Jumlah Penduduk')
plt.title('Jumlah Penduduk Laki-Laki dan Perempuan di Setiap Kecamatan pada Tahun 2020')
plt.xticks(x, data['Kecamatan'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Visualisasi jumlah penduduk laki-laki dan perempuan di setiap kecamatan pada tahun 2021
plt.figure(figsize=(12, 6))
plt.bar(x, data['2021_Laki-Laki'], width=bar_width, label='Laki-Laki 2021')
plt.bar(x, data['2021_Perempuan'], width=bar_width, bottom=data['2021_Laki-Laki'], label='Perempuan 2021')

plt.xlabel('Kecamatan')
plt.ylabel('Jumlah Penduduk')
plt.title('Jumlah Penduduk Laki-Laki dan Perempuan di Setiap Kecamatan pada Tahun 2021')
plt.xticks(x, data['Kecamatan'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Visualisasi perubahan jumlah penduduk di setiap kecamatan dari tahun 2020 ke tahun 2021
plt.figure(figsize=(12, 6))
change_laki_laki = data['2021_Laki-Laki'] - data['2020_Laki-Laki']
change_perempuan = data['2021_Perempuan'] - data['2020_Perempuan']

plt.bar(x, change_laki_laki, width=bar_width, label='Perubahan Laki-Laki')
plt.bar(x, change_perempuan, width=bar_width, bottom=change_laki_laki, label='Perubahan Perempuan')

plt.xlabel('Kecamatan')
plt.ylabel('Perubahan Jumlah Penduduk')
plt.title('Perubahan Jumlah Penduduk di Setiap Kecamatan dari Tahun 2020 ke Tahun 2021')
plt.xticks(x, data['Kecamatan'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Visualisasi total jumlah penduduk di setiap kecamatan pada tahun 2020 dan 2021
plt.figure(figsize=(12, 6))
total_2020 = data['2020_Jumlah']
total_2021 = data['2021_Jumlah']

x = range(len(data['Kecamatan']))
width = bar_width

plt.bar(x, total_2020, width=width, label='Total 2020')
plt.bar(x, total_2021, width=width, bottom=total_2020, label='Total 2021')

plt.xlabel('Kecamatan')
plt.ylabel('Total Jumlah Penduduk')
plt.title('Total Jumlah Penduduk di Setiap Kecamatan pada Tahun 2020 dan 2021')
plt.xticks(x, data['Kecamatan'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
